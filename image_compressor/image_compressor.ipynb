```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("random_image.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap="gray")
plt.title('Random Image')
plt.show()
```

```python
U, S, Vt = np.linalg.svd(image)

U.shape, S.shape, Vt.shape
```

```python
# First and last 10 singular values
S[:10], S[-10:]
```

```python
reconstructed_image = np.matrix(U[:,:1]) * np.diag(S[:1]) * np.matrix(Vt[:1,:])
plt.imshow(reconstructed_image, cmap="gray")
plt.title("Recontructed image with the first Single value")
plt.show()
```
