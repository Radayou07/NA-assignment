```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import sys
import os
```

```python
image = cv2.imread("random_image.jpg")
size = os.path.getsize("random_image.jpg") / 1024

plt.imshow(image)
plt.title(f'Random Image, with size: {size:.2f} KB')
plt.show()
```

```python
U, S, Vt = np.linalg.svd(image)

U.shape, S.shape, Vt.shape
```

```python
image_gray = cv2.imread("random_image.jpg", cv2.IMREAD_GRAYSCALE)
U_g, S_g, Vt_g = np.linalg.svd(image_gray)
reconstructed_gray_image = np.matrix(U_g[:,:300]) * np.diag(S_g[:300]) * np.matrix(Vt_g[:300,:])
plt.imshow(reconstructed_gray_image, cmap="gray")
plt.title(f"Recontructed image (Gray)")
plt.show()
```

```python
idk = [1, 10, 20, 50, 70, 100]
for i in idk:
    reconstructed_image = np.matrix(U_g[:,:i]) * np.diag(S_g[:i]) * np.matrix(Vt_g[:i,:])
    plt.imsave(f"./{i}.jpg", reconstructed_image, cmap="gray")
    size = os.path.getsize(f"{i}.jpg") / 1024
    plt.imshow(reconstructed_image, cmap="gray")
    plt.title(f"Recontructed image with the size {size:.2f} KB")
    plt.show()
```

```python
color_image = cv2.imread('random_image.jpg')
B, G, R = cv2.split(color_image)

plt.subplot(1, 3, 1)
plt.imshow(R, cmap='Reds_r')
plt.subplot(1, 3, 2)
plt.imshow(B, cmap='Blues_r')
plt.subplot(1, 3, 3)
plt.imshow(G, cmap='Greens_r')
plt.show()
```
